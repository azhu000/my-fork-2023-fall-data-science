It looks like my prediction was incorrect as it seems like it usually takes 40 wins for a team to make playoffs.

````python
independent_variables = 'point_diff'
dependent_variable = 'W'

X = df[independent_variables]
y = df[dependent_variable]

X = sm.add_constant(X)

model_wins = sm.OLS(y, X).fit()
y_pred = model_wins.predict(X) 

results = model_wins.summary()
print(results)

model_wins_root_mean_squared_error = rmse(y, y_pred)
print('#'*79)
print( 'Dependent Variable %s average: ' % dependent_variable, y.mean())
print( 'RMSE:', model_wins_root_mean_squared_error)
try:
    print( 'RMSEs Percentage off average:', ((model_wins_root_mean_squared_error / y.mean())*100).round(2))
except:
    pass
````

R2 and RSME Scores 

````python
# this is  a linear regression on 2PA, 3PA, FTA with PTS
independent_variables =["2PA", "3PA", "FTA"]
dependent_variable = 'PTS'

X = df[independent_variables]
y = df[dependent_variable]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model_1 = LinearRegression()

model_1.fit(X_train,y_train)

y_pred = model_1.predict(X_test)
print('R2 Score: ', r2_score(y_test,y_pred))

print("RMSE Score: ", mean_squared_error(y_test, y_pred, squared=False).round(5))
````

A R2 score of 0.78 indicates an okay fit (Note these are the attempts) however the RMSE score is rather high indicating the model isn't the best

````python
# LG on FG, 2P, and FT with PTS
independent_variables = ['FG','2P','FT']
dependent_variable = 'PTS'

X = df[independent_variables]
y = df[dependent_variable]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model_3 = LinearRegression()

model_3.fit(X_train,y_train)

y_pred = model_3.predict(X_test)
print('R2 Score: ', r2_score(y_test,y_pred))

print("RMSE Score: ", mean_squared_error(y_test, y_pred, squared=False))
````

With an R2 score of 1 and a RMSE score of 1.148, this model is much better than the previous one with attempted shots
